```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
data_raw = pd.read_csv('China_clean.csv')
```

```python
years = data_raw.columns[1:]
```

```python
indexes = data_raw['year']
```

```python
data_raw.loc[0][1:]
```

```python
data_clean = pd.DataFrame()
data_clean['years'] = years
data_clean.shape
```

```python
for it in range(len(indexes)):
    data = np.array(data_raw.loc[it][1:])
    label = data_raw.loc[it][0]
    data_clean[label]=data
```

```python
data_clean
```

```python
data_new = data_clean
from sklearn import linear_model
```

```python
for col in data_new.columns[1:]:
    notnull=(1-data_new[col].isnull()).astype(bool)
    print(notnull.sum())
    regr = linear_model.LinearRegression()
    years_available = data_new.years[notnull].reshape(-1,1)
    y = data_new[col][notnull]
    regr.fit(years_available,y)
    
    null = (1-notnull).astype(bool)
    years_null = data_new.years[null].reshape(-1,1)
    if len(years_null)==0:
        continue
    missing = regr.predict(years_null)
    
    data_new[col][null] = missing
```
