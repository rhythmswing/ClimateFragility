```python
import csv
```

```python
def input_data():
    csvFile = open("modified_fsi-2017.csv", "r")
    reader = csv.reader(csvFile)
    attr_dic = {}
    score_dic = {}
    for item in reader:
        # ignore the first line
        if reader.line_num == 1:
            continue
        attr_dic[item[0]] = [float(i) for i in item[4:]]
        score_dic[item[0]] = float(item[3])
    csvFile.close()
```

```python
input_data()
```
